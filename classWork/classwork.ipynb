Q. Fibonacci series using recursion

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

number = int(input("Enter number: "))
if number > 0:
    print("Fibonacci series:")
    for i in range(number):
        print(fibonacci(i), end=" ")
else:
    print("Please enter a positive number.")
    
```

Use of class and methods

```python
class Animal:
    def __init__(self,**kwargs):
        self._type = kwargs.get('type', 'kitten')
        self._name = kwargs.get('name', 'fluffy') 
        self._sound = kwargs.get('sound', 'meow')

    def type(self, t = None):
        if t: self._type = t
        return self._type

    def name(self, n = None):
        if n: self._name = n
        return self._name

    def sound(self, s = None):
        if s: self._sound = s
        return self._sound

    def __str__(self):
        return f"{self._name} is a {self._type} that says {self._sound}"
        
def main():
    a0 = Animal(type='kitten', name='fluffy', sound='meow')
    a1 = Animal(type='puppy', name='fido', sound='woof')
    print(a0)
    print(a1)
if __name__ == "__main__": main()
```

Use class and inheritance

```python
class Animal:
    def __init__(self, **kwargs):
        if 'type' in kwargs: self._type = kwargs['type']
        if 'name' in kwargs: self._name = kwargs['name']
        if 'sound' in kwargs: self._sound = kwargs['sound']
        
    def type(self, t = None):
        if t: self._type = t
        try: return self._type
        except AttributeError: return None
        
    def name(self, n = None):
        if n: self._name = n
        try: return self._name
        except AttributeError: return None
        
    def sound(self, s = None):
        if s: self._sound = s
        try: return self._sound
        except AttributeError: return None
        
class Duck(Animal):
    def __init__(self, **kwargs):
        self._type = 'duck'
        if 'type' in kwargs: self._type = kwargs['type']
        super().__init__(**kwargs)
        
class kitten(Animal):
    def __init__(self, **kwargs):
        self._type = 'kitten'
        if 'type' in kwargs: self._type = kwargs['type']
        super().__init__(**kwargs)
        
def print_animal(o):
    if not isinstance(o, Animal):
        raise TypeError("Expected an Animal object")
    print(f'The {o.name()} is a {o.type()} that says {o.sound()}')
    
def main():
    d = Duck(name='Donald', sound='Quack Quack')
    k = kitten(name='Fluffy', sound='Meow Meow')
    print_animal(d)
    print_animal(k)
if __name__ == "__main__": main()
```

Q. Take name from the user and in output give repeted letter.

```python
name_input = input("Enter a name: ")
char_counts = {}
repeated_letters = []

name_lower = name_input.lower()

for char in name_lower:
    if 'a' <= char <= 'z':
        char_counts[char] = char_counts.get(char, 0) + 1

for char, count in char_counts.items():
    if count > 1:
        repeated_letters.append(char)

if repeated_letters:
    print(f"Repeated letters in '{name_input}': {', '.join(repeated_letters)}")
else:
    print(f"No repeated letters found in '{name_input}'.")
```
